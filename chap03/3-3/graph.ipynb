```python
import numpy as np
import matplotlib.pyplot as plt

# L1  L2  
l1 = np.linspace(-10, 10, 400)
l2 = np.linspace(-10, 10, 400)
L1, L2 = np.meshgrid(l1, l2)
L1_penalty = np.abs(L1) + np.abs(L2)
L2_penalty = L1**2 + L2**2

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# L1  
ax[0].contour(L1, L2, L1_penalty, levels=20)
ax[0].set_title('L1 Regularization (Lasso)')
ax[0].set_xlabel('Coefficient 1')
ax[0].set_ylabel('Coefficient 2')

# L2  
ax[1].contour(L1, L2, L2_penalty, levels=20)
ax[1].set_title('L2 Regularization (Ridge)')
ax[1].set_xlabel('Coefficient 1')
ax[1].set_ylabel('Coefficient 2')

plt.show()
plt.savefig('regularization.png')
```
